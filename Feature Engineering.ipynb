# Removing NAs

```python
def remove_na(col):
    #print(col)
    means = final.groupby(["product_type", "sub_area","ecology"])[col].agg({lambda x: x.mean(skipna=True)})
    
    df = final[[col,"product_type","sub_area","ecology"]]
    df[col] =  df[col].fillna(0.0)
    for i in range(0, len(df[col])):
        if df[col][i] == 0.0:
            index1 = df["product_type"][i]
            index2 = df["sub_area"][i]
            index3 = df["ecology"][i]
            df[col][i] = means.loc[index1,index2,index3]
    final[col] = df[col]
```

# Picking top 50 columns to retain

```python
columns = final.columns.drop(["price_doc"])
corrs = {}
price = final["price_doc"]
for col in columns:
    corr = price.corr(final[col])
    corrs[col] = corr
    corrs[col] = np.abs(corrs[col])
```

```python
corrs = sorted(corrs.items(), key = lambda x: x[1],reverse = True)
corrs
```

```python
macros = macro.columns
final_columns = []
mac = 0
total = 0
for cor in corrs:
    if cor[0] in macro:
        if mac < 10 and total < 50:
            final_columns.append(cor[0])
            mac+=1
            total += 1
    elif total < 50:
        final_columns.append(cor[0])
        total +=1
#final_columns
```

# Final Dataset

```python
final = final[final_columns]
final["price_doc"]  = price
final
```
